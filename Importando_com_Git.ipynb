# Import unitário Github RAW

```python
# ------- para deixar a tabela do json mais bonita visualmente -------
import pandas as pd
# requests: pega um valor da web
import requests

# importar o arquivo json para o sistema
url_raw = input('Cole o link do json aqui: ')
resposta = requests.get(url_raw)
# comando: vá até o github e pegue o conteúdo da URL
dados = resposta.json()

df_json = pd.DataFrame(dados)
# agora vamos arrumar os dados para uma tabela

# comando try: parecido com o if, só que ele tenta até dar erro
try:
  nome_tabela = input('Informe o nome da tabela: ')
  tabela_dados = pd.json_normalize(dados[nome_tabela]) # o nome da tabela tem q ser igual do arquivo
  display(tabela_dados)

except KeyError:
  print('Não consigo ler o arquivo')
```

# Verificação dos dados

```python
# Padronização das informações
tabela_dados['department.name'] = tabela_dados['department.name'].str.strip().str.upper()

# Para ver
print(tabela_dados['department.name'])

# Limpeza
# Apagar os dados que não quer exibir
tabela_dados = tabela_dados.drop(columns='department.blablabla')
```

```python
# Renomear campos
tabela_dados = tabela_dados.rename(columns={
    "name": "Nome",
    "department.name": "Departamento",
    "department.role": "Função"
})

display(tabela_dados)
```
