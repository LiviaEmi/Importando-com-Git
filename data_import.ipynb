```python
# importando dados do github

# 1ª maneira - baixar um repo para o colab
# !git clone <html>
# não é indicado, pq ele copia tudo do github

# 2ª maneira - importação via RAW
# importação unitária dos dados
# vai no github, abre cada arquivo, aperta em 'raw' e copia o link da página
# para arquiv do tipo xlsx, não precisa abrir o raw, só clica no arquivo >
# botão direito cabeçalho da pag > copiar link (arquivo 3)

arquivo1 = 'https://raw.githubusercontent.com/LiviaEmi/Importando-dados-com-Python/refs/heads/main/example_data.csv'
arquivo2 = 'https://raw.githubusercontent.com/LiviaEmi/Importando-dados-com-Python/refs/heads/main/example_data.json'
arquivo3 = 'https://github.com/LiviaEmi/Importando-dados-com-Python/raw/refs/heads/main/example_data.xlsx'
arquivo4 = 'https://raw.githubusercontent.com/LiviaEmi/Importando-dados-com-Python/refs/heads/main/sample_data.csv'

# para os csv - importar pandas
import pandas as pd
df_csv_exemple = pd.read_csv(arquivo1)
df_csv_sample = pd.read_csv(arquivo4)

# para o json
df_example_data = pd.read_json(arquivo2)
display(df_example_data)

# para o excel
# precisamos usar engine='openpyxl'
df_example_data_excel = pd.read_excel(arquivo3, engine='openpyxl')
display(df_example_data_excel)
```
